```python
# Stree When Depth is Constant
Q = int(input("Enter the value of Laod in kN :"))
N = int(input("Number of Data values of Radial distance :"))
Z = int(input("Enter the value of Depth :"))
r = []
for i in range (1,N+1):
  print("Enter radial distance in m".format(i))
  Value_r = float(input())
  r.append(Value_r)
  Stress = round((((3*Q)/(2*3.14*Z*Z))*(((1/(1+((Value_r/Z)**2))))**2.5)),3)
  print("Stress =",Stress,"kN/m2")
```

```python
# Stress when Radius is Constant
Q = int(input("Enter the value of Laod in kN :"))
M = int(input("Number of Data values of Depth :"))
r = int(input("Enter the value of Radial distance :"))
pi = 3.14159265359
Z = []
for j in range (1,M+1):
  print("Enter Depth".format(j))
  Value_Z = float(input())
  Z.append(Value_Z)
  Stress = round((((3*Q)/(2*pi*Value_Z*Value_Z))*(((1/(1+((r/Value_Z)**2))))**2.5)),3)
  print("Stress =",Stress,"kN/m2")
```
