```python
#Design Concrete mix as per IS 10262
fck = float(input("Enter the value of Characteristic compressive strength :"))

# Experimental Determinations
Gca = float(input("Enter the value of specific gravity of CA :"))
Gfa = float(input("Enter the value of specific gravity of FA :"))
Gc = float(input("Enter the value of specific gravity of Cement :"))
Water_Density = float(input("Enter the value of Water Density :"))
AGG_Size = float(input("Enter the nominal Size of Aggregate :"))
Nature_of_AGG = input("Enter the nature of Aggregate (Sub-Angular/Gravel/Round) :")
Slump = float(input("Enter the Workability of concrete :"))
Admixture = input("Type of Admixture (Plastisizer/Super-plastisizer) :")
Exposure_Condition = input("Exposure Condition (Mild/Moderate/Severe/Very Severe/Extreme) :")
Concreting = input("Type of Concreting (Plain/Rainforced) :")
Zone = int(input("Zone (1/2/3/4) :"))

#Target Mean Strength
sigma = {
    10:3.5,
    15:3.5,
    20:4,
    25:4,
    30:5,
    35:5,
    40:5,
    45:5,
    50:5,
}
if fck not in sigma:
  raise ValueError("Invalid Characteristic compressive strength")

ft = fck+sigma[fck]*1.65
ft = round (ft,3)
print("Target Mean Strength :",ft,"Mpa")

#Maximum free Water Cement Ratio
#Refer IS 456:2000 Table 5
if(Concreting == "Plain"):
  WC_ratio ={
      "Mild" : 0.6,
      "Moderate" : 0.6,
      "Severe" : 0.5,
      "Very Sever" : 0.45,
      "Extreme" : 0.4,
  }
else:
  WC_ratio ={
      "Mild" : 0.55,
      "Moderate" : 0.5,
      "Severe" : 0.45,
      "Very Sever" : 0.45,
      "Extreme" : 0.4,
  }

WC_ratio = WC_ratio[Exposure_Condition]
WC_ratio = round (WC_ratio,3)
print("Maximum free Water Cement Ratio :",WC_ratio)

#Minimum Cement Content
if(Concreting == "Plain"):
  Min_Cement_Content = {
      "Mild" : 220,
      "Moderate" : 240,
      "Severe" : 250,
      "Very Sever" : 260,
      "Extreme" : 280,
  }
else:
  Min_Cement_Content = {
      "Mild" : 300,
      "Moderate" : 300,
      "Severe" : 320,
      "Very Sever" : 340,
      "Extreme" : 360,
  }

Min_Cement_Content = Min_Cement_Content[Exposure_Condition]
Min_Cement_Content = round (Min_Cement_Content,3)
print("Minimum Cement Content :",Min_Cement_Content,"kg/m3")

#Water Content
Water_Content = {
    10:208,
    20:186,
    40:165,
}
Water_Content = Water_Content[AGG_Size]
if(Slump == 75):
  Water_Content = Water_Content+Water_Content*0.03
elif(Slump == 100):
  Water_Content = Water_Content+Water_Content*0.06
elif(Slump == 125):
  Water_Content = Water_Content+Water_Content*0.9
elif(Slump == 150):
  Water_Content = Water_Content+Water_Content*0.12
elif(Slump == 175):
  Water_Content = Water_Content+Water_Content*0.15
elif(Slump == 200):
  Water_Content = Water_Content+Water_Content*0.18

if(Nature_of_AGG == "Sub-Angular"):
  Water_Content = Water_Content-10
elif(Nature_of_AGG == "Gravel"):
  Water_Content = Water_Content-20
elif(Nature_of_AGG == "Round"):
  Water_Content = Water_Content-25

if(Admixture == "Plastisizer"):
  Water_Content = Water_Content-(0.1*Water_Content)
elif(Admixture == "Super-plastisizer"):
  Water_Content = Water_Content-(0.2*Water_Content)

Water_Content = round (Water_Content,3)
print("Water Content :",Water_Content,"kg/m3")

#Cement Content
Cement_Content = Water_Content/WC_ratio
Cement_Content = round (Cement_Content,3)
print("Cement Content :",Cement_Content,"kg/m3")
print("As per IS 456:2000, Maximum allowed Cement Content is 450 kg/m3")
if(Cement_Content < 450):
  Cement_Content = Cement_Content
else:
  Cement_Content = 450
if Cement_Content < 450:
  print("Safe")

#Valume Calculation
Vol_Cement = Cement_Content/(Gc*Water_Density)
Vol_Cement = round (Vol_Cement,3)
print("Cement Volume :",Vol_Cement,"m3")

Vol_Water = Water_Content/Water_Density
Vol_Water = round (Vol_Water,3)
print("Water Volume :",Vol_Water,"m3")

Vol_AGG = 1- Vol_Water - Vol_Cement
Vol_AGG = round (Vol_AGG,3)
print("Volume of FA & CA:",Vol_AGG,"m3")

Zone_ID = {}
Zone_ID[1]={10:0.44,20:0.60,40:0.69}
Zone_ID[2]={10:0.46,20:0.62,40:0.71}
Zone_ID[3]={10:0.48,20:0.64,40:0.73}
Zone_ID[4]={10:0.5,20:0.66,40:0.75}

Fraction = Zone_ID[Zone][AGG_Size]

if(WC_ratio == 0.5):
  Fraction=Fraction
elif(WC_ratio == 0.45):
  Fraction=Fraction+(0.01*Fraction)
elif(WC_ratio == 0.4):
  Fraction=Fraction+(0.02*Fraction)
elif(WC_ratio == 0.55):
  Fraction=Fraction+(0.01*Fraction)
elif(WC_ratio == 0.60):
  Fraction=Fraction+(0.02*Fraction)

Fraction = round(Fraction,3)
print("Cource Aggregate Fraction :",Fraction)

Vol_CA = Vol_AGG*Fraction
Vol_CA = round(Vol_CA,3)
print("Volume of Coarce Aggregate :",Vol_CA,"m3")

Vol_FA = Vol_AGG-Vol_CA
Vol_FA = round(Vol_FA,3)
print("Volume of Fine Aggregate :",Vol_FA,"m3")

Mass_CA = Vol_CA*Gca*Water_Density
Mass_CA = round(Mass_CA,3)
print("Mass of Cource Aggregate :",Mass_CA,"kg/m3")

Mass_FA = Vol_FA*Gfa*Water_Density
Mass_FA = round(Mass_FA,3)
print("Mass of Fine Aggregate :",Mass_FA,"kg/m3")

#Ratios
print("Weight Batching")
WB1 = Cement_Content/Cement_Content
WB1 = round(WB1,3)

WB2 = Mass_FA/Cement_Content
WB2 = round(WB2,3)

WB3 = Mass_CA/Cement_Content
WB3 = round(WB3,3)

WB4 = Water_Content/Cement_Content
WB4 = round(WB4,3)

print(WB1,":",WB2,":",WB3,":",WB4)

print("Volume Batching")
VB1 = Vol_Cement/Vol_Cement
VB1 = round(VB1,3)

VB2 = Vol_FA/Vol_Cement
VB2 = round(VB2,3)

VB3 = Vol_CA/Vol_Cement
VB3 = round(VB3,3)

VB4 = Vol_Water/Vol_Cement
VB4 = round(VB4,3)

print(VB1,":",VB2,":",VB3,":",VB4)
```
