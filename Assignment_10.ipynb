```python
# Design of tension member
tu = float(input("Enter the value of ultimate tensile strength : "))
fy = float(input("Enter the value of yield strength of steel : "))
fu = float(input("Enter the value of ultimate strength of steel : "))
fub = float(input("Enter the value of ultimate strength of bolt : "))
Gamma_mo = float(input("Enter the value of partial factor of safety Garmma_mo : "))
Gamma_mi = float(input("Enter the value of partial factor of safety Garmma_m1 : "))
Gamma_mb = float(input("Enter the value of partial factor of safety Gamma_ mb : "))
print ("Gross Area Required")
Agreq = 1.1* tu* 1000/fy
print ("The value of gross area required is : ", round((1.2*Agreq),3))

# Selection of section
# Selecting ISA 100x65x8
Ag = float(input("Enter the value of gross areaof steel is : "))
Lcl = float(input("Enter the length of connected leg : "))
Lol = float(input("Enter the length of outstand leg : "))
t = float(input("Entert the value of least thickness : "))
Ag = 1257

# Design of connections
d = float(input("Enter the value of diameter of bolt : "))
do = d+2
print ("The diameter of bolt hole is : ", round(do,3))

# As per IS code minimum pitch distance is
pin = 2.5*d
print ("The minimum pitch is : ", round(pin,3))

# Edge distance as per IS 800 is
e = 1.5*do
print ("Enter the value of edge distance : ", round(e,3))

# print ("threaded area of bolt is:", Anb)
Asb = 0.7854*d*d
print ("plane shank area of bolt is : ", Asb)
nn = float(input("Number of shear planes with threaded intercepting the shear plane : "))
ns = float(input("Number of shear plane without threads : "))
Anb = 0.78*0.7854*d*d
Vdsb = ((fub/(1.732* Gamma_mb)*(nn* Anb+ ns*Asb))/1000)
print ("The value of Vdsb :", round(Vdsb,3))
kbl = e/(3*do)
print ("Kbl :", round(kbl,3))
kb2 = (pin/(3*do))- 0.25
print ("Kb2 :", round(kb2,3))
kb3 = fub/fu
print ("Kb3 :", round(kb3,3))
kb4 = 1
print ("Kb4 :", round(kb4,3))
kb = min(kbl, kb2, kb3, kb4)
print ("Kb :", round(kb,3))
Vdpb = (2.5 * kb*d*t*fu*10**-3)/Gamma_mb
print ("Vdpb :", round(Vdpb,3))
vd = min(Vdsb, Vdpb)
print ("Vd :", round(vd,3))
N = tu/vd
print ("Number of bolts requird :", round(N,3))
N = float(input("Enter the value of number of bolts :"))

#Check for strength
# Criteria 1 Yeilding of Gross Section
Tdg = (Ag*fy*10**-3)/Gamma_mo
print ("The value of tensile strength due to yielding of gross section is :", round(Tdg,3))

# Criteria 2 Rupture
Anc = (Lcl-(t/2)-do)*t
print ("Net Area of Connecting leg is: (Anc) :", round(Anc,3))
Ago = (Lol-(t/2))*t
print ("Gross Area of outstand leg is: (Anc) :", round(Ago,3))
Lc = (N-1)*pin
print ("Le :", round(Lc,3))
bs = 0.6*Lcl+ Lol*t
print ("bs :", round(bs,3))
Beta = (fy/fu)*(bs/Lc)*(Lol/t)*1.4+(0.076*(fy/tu)*(bs/Lc)*(Lol/t))
print ("Beta :", round(Beta,3))
print ("Check 1")
if Beta>1.4:
 print ("Not Safe")
else:
 print ("Safe")
print ("Check 2")
if Beta<0.7:
 print ("Not Safe")
else:
 print ("Safe")
Tdn = ((0.9 * fu*Anc)/Gamma_mi + (Beta *Ago*fy/Gamma_mo))/1000
print ("Tdn :", round(Tdn,3))

# Criteria 3 Block Shear
Avg = (pin* (N-1)+e)*t
print ("Avg :", round(Avg,3))
Avn = ((pin*(N-1) +e)-(N-1)*do+(8.5*do))*t
print ("Avn :", round(Avn,3))
Atg = 0.6* Lcl *t
print ("Atg :", round(Atg,3))
Atn = Atg*0.5*do
print ("Atn :", round(Atn,3))
Tb1 = (((Avg*fy)/(1.732 *Gamma_mo))+(0.9*fu*Atn)/Gamma_mi)/1000
print ("Tb1 :", round(Tb1,3))
Tb2 = ((((0.9*Avn*fu)/(1.732* Gamma_mi) + (Atg*fy)/Gamma_mo)))/1000
print ("Tb2 :", round(Tb2,3))
Tb = min (Tb1, Tb2)
print ("Tb :", round(Tb,3))
Td = min(Tdg,Tdn,Tb)
print ("Td :", round(Td,3))

# Corrected indentation and comparison
if Td > tu: # Assuming 'tu' is the ultimate tensile strength to compare against
 print ("SAFE")
else:
 print ("Revise the Section")
```
